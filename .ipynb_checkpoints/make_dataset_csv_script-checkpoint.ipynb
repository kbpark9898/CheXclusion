```python
import pandas as pd
import time
```

```python
data = pd.read_csv("./Data_Entry_2017_v2020.csv")
data['Finding Labels']=="Cardiomegaly"
```

```python
data_test = pd.read_csv("./test_list.txt")
test_set=[]
for i in data_test['Image Index']:
    test_set.append(i)
```

```python
data_train_valid = pd.read_csv("./train_val_list.txt")
data_train_valid
```

```python
index=0
valid_set =[]
train_set =[]
for i in data_train_valid["Image Index"]:
    if index%10 ==0:
        valid_set.append(i)
    else:
        train_set.append(i)
    index+=1
print(len(valid_set), len(train_set))
```

```python
col_name=[]
for col in data.columns:
    col_name.append(col)
print(col_name)
```

```python

for i in range(len(test_set)):
    print("test : ", i)
    df_mask=data['Image Index'] == test_set[i]
    current_df = data[df_mask]
    test_df=test_df.append(current_df, ignore_index=True)
test_df.to_csv("test_df.csv")
```

```python
start = time.time()
for i in range(len(train_set)):
    print("train : ", i)
    df_mask=data['Image Index'] == train_set[i]
    current_df = data[df_mask]
    train_df=train_df.append(current_df, ignore_index=True)    
train_df.to_csv("train_df.csv")
print("end: ", start-time.time())
```
